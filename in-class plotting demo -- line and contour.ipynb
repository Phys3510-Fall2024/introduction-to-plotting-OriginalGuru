## Plotting with `matplotlib`

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the potential energy
def potential_energy(x, k=1.0):
    return 0.5 * k * x**2 # k is the spring constant

# test function for the potential energy
potential_energy(2,10)
```

```python
# define mesh for plotting
x = np.linspace(-10, 10, 100) # np.linspace(x_min, x_max, npts)

# define y values from x list
y = potential_energy(x)

# testing the output
# print(np.transpose([x, y]))
```

```python
# plot the function
plt.figure(figsize=(4,3)) # figure size in inches
plt.plot(x, y, color="black", linewidth=0.5)

# add axes labels
plt.xlabel("displacement (arbitrary unit)")
plt.ylabel("potential energy (arbitrary units)")

# change limits of x and y
plt.xlim([-4, 4])
plt.ylim([-1, 10])

plt.show() # combine all elements of the plot
```

## Contour plots

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the potential energy
def potential_energy_2D(x, y, kx=1.0, ky=1.0, c=.25):
    return 0.5 * (kx * x**2 + ky * y**2) + c * x * y**2 # (kx, ky) are the spring constants

# test function for the potential energy
potential_energy_2D(2,1)
```

```python
# define meshes for plotting
x = np.linspace(-5, 5, 100) # np.linspace(x_min, x_max, npts)
y = np.linspace(-5, 5, 100) # np.linspace(x_min, x_max, npts)

# combine 1D meshes into meshgrid
X, Y = np.meshgrid(x,y)

# define Z values
Z = potential_energy_2D(X,Y)

# testing the mesh
# print(X)
# print()
# print(Y)
# print()
# print(Z)
```

```python
# construct the contour plot
plt.figure(figsize=(4,3)) # figure size in inches
plt.contourf(X, Y, Z, levels=30)

# add a colorbar legend
plt.colorbar(label='potential energy (arbitrary units)')

# add axes labels
plt.xlabel("x-displacement (arbitrary unit)")
plt.ylabel("y-displacement (arbitrary unit)")

# change limits of x and y
plt.xlim([-4, 4])
plt.ylim([-4, 4])

plt.show() # combine all plot objects
```
